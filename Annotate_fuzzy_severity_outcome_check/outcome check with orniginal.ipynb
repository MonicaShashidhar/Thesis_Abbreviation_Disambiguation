```python
from tqdm import tqdm

count_true = 0
true_positive = 0  # case predicted as survivor and in actual the case is survivor
false_negative = 0  # case predicted as survivor and in actual the case is dead
true_negative = 0  # case predicted as death and in actual the case is death
false_positive = 0  # case predicted as death and in actual the case is survivor

for index in tqdm(range(len(df[:200]))):
    if pd.notna(df.iloc[index].manual_outcome) and pd.notna(df.iloc[index].Annotated_outcome_redefined):
        if df.iloc[index].manual_outcome == df.iloc[index].Annotated_outcome_redefined:
            count_true += 1

        if df.iloc[index].manual_outcome == "survivor" and df.iloc[index].Annotated_outcome_redefined == "survivor":
            true_positive += 1
        elif df.iloc[index].manual_outcome == "survivor" and df.iloc[index].Annotated_outcome_redefined == "death":
            false_negative += 1
        elif df.iloc[index].manual_outcome == "death" and df.iloc[index].Annotated_outcome_redefined == "death":
            true_negative += 1
        elif df.iloc[index].manual_outcome == "death" and df.iloc[index].Annotated_outcome_redefined == "survivor":
            false_positive += 1
```

```python
# survivor
Precision = true_positive / (true_positive + false_positive)
Recall = true_positive / (true_positive + false_negative)
F1_Score = 2 * ((Precision*Recall)/(Precision+Recall))
```

```python
print("Accuracy: ",count_true/len(df[:200]))
print("Precision: ",Precision)
print("Recall: ",Recall)
print("F1 Score: ",F1_Score)
```
