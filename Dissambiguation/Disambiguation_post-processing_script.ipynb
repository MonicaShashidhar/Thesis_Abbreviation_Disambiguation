```python
df = pd.read_csv("/content/drive/My Drive/MACROBATT/FINAL_SURVIVOR_DEATH/DISSAMBIGUATION/dissambiguation_data.csv")
```

```python
def process_dict(temp_result):
    final_dict = {}
    for key in temp_result:
        if type(temp_result[key])==list:
            temp_list = temp_result[key]
            all_matching_list = []
            for data in temp_list:
                if len(key)==len(data[0].split()):
                    if all([True  if i.lower()==j[0].lower() else False for i,j in zip(key,data[0].split())]):
                        all_matching_list.append(data)


#             print("All matching results: ",all_matching_list)
            if len(all_matching_list)==0:
                final_dict[key] = temp_list[0][0]
            else:
                final_dict[key] = sorted(all_matching_list,key=lambda x:x[1],reverse=True)[0][0]

        else:
            final_dict[key] = temp_result[key]

    return final_dict


```

```python
temp_result = eval(df.iloc[0]['Results'])
process_dict(temp_result)
# temp_result
```

```python
replaced_text=[]
for index in range(len(df)):

    temp_result = eval(df.iloc[index]['Results'])
    final_dict = process_dict(temp_result)
    if final_dict:
        final_string  = df.iloc[index]['Text'].replace("\n"," ")
        for key in final_dict:
            final_string = final_string.replace(" "+key," "+final_dict[key])
        replaced_text.append(final_string)
    else:
        replaced_text.append(df.iloc[index]['Text'].replace("\n"," "))


# df.iloc[0]['Text'].replace("\n"," ")
```

```python
df['Replaced_Text'] = replaced_text
```

```python
df.to_csv("/content/drive/My Drive/MACROBATT/FINAL_SURVIVOR_DEATH/DISSAMBIGUATION/new_disambiguation_data.csv",index=False)
```
